### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
# Example DataFrame (replace this with your actual dataset)
df = pd.DataFrame({
    'Age': ['18-24 years old', '25-34 years old', '25-34 years old', '35-44 years old', '18-24 years old', '25-34 years old', '35-44 years old'],
    'SOPartFreq': ['Daily', 'Weekly', 'Monthly or less', 'Daily', 'Weekly', 'Daily', 'Monthly or less']
})

# Step 1: Group by Age and SOPartFreq, then count occurrences
grouped = df.groupby(['Age', 'SOPartFreq']).size().reset_index(name='Count')

# Step 2: Create bubble plot
plt.figure(figsize=(12, 6))

scatter = plt.scatter(
    x=grouped['Age'],
    y=grouped['SOPartFreq'],
    s=grouped['Count'] * 100,   # Bubble size scaled
    alpha=0.6,
    edgecolors='black',
    c=grouped['Count'],         # Optional: color by count
    cmap='viridis'
)

# Add color bar for context
cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.title('Bubble Plot: Age Group vs. Frequency of Participation (SOPartFreq)')
plt.xlabel('Age Group')
plt.ylabel('Participation Frequency (SOPartFreq)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
# Example data (replace with your actual DataFrame)
df = pd.DataFrame({
    'CompTotal': [50000, 80000, 30000, 100000, 60000, 40000, 120000],
    'JobSatPoints_6': [3, 4, 2, 5, 3, 2, 4]
})

# Step 1: Drop missing or invalid values
df = df.dropna(subset=['CompTotal', 'JobSatPoints_6'])

# Optional: Normalize compensation if the range is too wide
# df['CompTotal'] = df['CompTotal'].clip(upper=200000)

# Step 2: Set a bubble size (can be based on another column, or static)
# For this case, we'll use a static size or based on compensation to illustrate
df['BubbleSize'] = df['CompTotal'] / 1000  # Scale down for visualization

# Step 3: Plot
plt.figure(figsize=(18, 8))
scatter = plt.scatter(
    x=df['CompTotal'],
    y=df['JobSatPoints_6'],
    s=df['BubbleSize'],  # Bubble size
    alpha=0.6,
    edgecolors='black',
    c=df['JobSatPoints_6'],  # Optional: color bubbles by job satisfaction
    cmap='coolwarm'
)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Job Satisfaction (Points)')

plt.title('Bubble Plot: Compensation vs. Job Satisfaction')
plt.xlabel('Total Compensation')
plt.ylabel('Job Satisfaction (Points)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here

# Example DataFrame (replace with your actual dataset)
df = pd.DataFrame({
    'Age': [
        '25-34 years old', '25-34 years old', '35-44 years old',
        '18-24 years old', '35-44 years old', '25-34 years old'
    ],
    'LanguageHaveWorkedWith': [
        'Python;JavaScript', 'Python;SQL', 'Java;SQL',
        'Python;C++', 'JavaScript;Python', 'C++;Python'
    ]
})

# Step 1: Explode LanguageHaveWorkedWith into multiple rows
df = df.dropna(subset=['Age', 'LanguageHaveWorkedWith']).copy()
df['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df.explode('LanguageHaveWorkedWith').rename(columns={'LanguageHaveWorkedWith': 'Language'})

# Step 2: Group by Age and Language, then count
grouped = df_exploded.groupby(['Age', 'Language']).size().reset_index(name='Count')

# Optional: Sort age categories for proper order
age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older'
]
grouped['Age'] = pd.Categorical(grouped['Age'], categories=age_order, ordered=True)
grouped = grouped.sort_values(['Age', 'Language'])

# Step 3: Create Bubble Plot
plt.figure(figsize=(12, 7))
scatter = plt.scatter(
    x=grouped['Age'],
    y=grouped['Language'],
    s=grouped['Count'] * 150,  # Bubble size scale
    alpha=0.6,
    edgecolors='black',
    c=grouped['Count'],
    cmap='plasma'
)

# Add color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.title('Bubble Plot: Programming Languages Used by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Programming Language')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here

# Sample data (replace with your actual DataFrame)
df = pd.DataFrame({
    'DatabaseWantToWorkWith': [
        'PostgreSQL;MySQL', 'MongoDB;PostgreSQL', 'MySQL', 'PostgreSQL;Oracle',
        'MongoDB', 'MySQL;SQLite', 'SQLite;MongoDB'
    ],
    'JobSatPoints_6': [4, 5, 3, 4, 2, 3, 4]
})

# Step 1: Explode database preferences into multiple rows
df = df.dropna(subset=['DatabaseWantToWorkWith', 'JobSatPoints_6']).copy()
df['DatabaseWantToWorkWith'] = df['DatabaseWantToWorkWith'].str.split(';')
df_exploded = df.explode('DatabaseWantToWorkWith').rename(columns={'DatabaseWantToWorkWith': 'Database'})

# Step 2: Group by Database and compute count and average satisfaction
grouped = df_exploded.groupby('Database').agg(
    RespondentCount=('Database', 'count'),
    AvgJobSat=('JobSatPoints_6', 'mean')
).reset_index()

# Step 3: Plot Bubble Plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(
    x=grouped['Database'],
    y=grouped['AvgJobSat'],
    s=grouped['RespondentCount'] * 150,  # Bubble size
    alpha=0.6,
    edgecolors='black',
    c=grouped['AvgJobSat'],  # Color by satisfaction
    cmap='coolwarm'
)

# Add color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Average Job Satisfaction')

plt.title('Bubble Plot: Preferred Databases vs. Job Satisfaction')
plt.xlabel('Preferred Database')
plt.ylabel('Average Job Satisfaction (Points)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here

# Sample DataFrame (replace with actual dataset)
df = pd.DataFrame({
    'DevType': [
        'Web developer;Mobile developer',
        'Data scientist or ML specialist',
        'Web developer',
        'Desktop applications developer',
        'Data scientist or ML specialist',
        'Database administrator;Data analyst'
    ],
    'ConvertedCompYearly': [70000, 120000, 80000, 65000, 115000, 90000],
    'JobSatPoints_6': [4, 5, 3, 2, 4, 3]
})

# Step 1: Explode DevType if multiple roles per respondent
df = df.dropna(subset=['DevType', 'ConvertedCompYearly', 'JobSatPoints_6']).copy()
df['DevType'] = df['DevType'].str.split(';')
df_exploded = df.explode('DevType')
df_exploded['DevType'] = df_exploded['DevType'].str.strip()

# Step 2: Group by DevType to get average compensation and job satisfaction
grouped = df_exploded.groupby('DevType').agg(
    AvgComp=('ConvertedCompYearly', 'mean'),
    AvgJobSat=('JobSatPoints_6', 'mean'),
    Count=('DevType', 'count')
).reset_index()

# Step 3: Create the bubble plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(
    x=grouped['DevType'],
    y=grouped['AvgComp'],
    s=grouped['AvgJobSat'] * 300,  # Bubble size scaled by satisfaction
    alpha=0.6,
    edgecolors='black',
    c=grouped['AvgJobSat'],  # Color by job satisfaction
    cmap='viridis'
)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Average Job Satisfaction')

# Customize plot
plt.title('Bubble Plot: Compensation by Developer Role (Bubble = Job Satisfaction)')
plt.xlabel('Developer Role')
plt.ylabel('Average Compensation (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here

# Sample DataFrame — replace this with your actual data
df = pd.DataFrame({
    'Age': [
        '25-34 years old', '25-34 years old', '35-44 years old',
        '18-24 years old', '35-44 years old', '25-34 years old'
    ],
    'NEWCollabToolsHaveWorkedWith': [
        'Slack;Zoom', 'Slack;Teams', 'Teams;Zoom',
        'Slack;Discord', 'Zoom;Slack', 'Teams;Slack'
    ]
})

# Step 1: Clean and split collaboration tools into rows
df = df.dropna(subset=['Age', 'NEWCollabToolsHaveWorkedWith']).copy()
df['NEWCollabToolsHaveWorkedWith'] = df['NEWCollabToolsHaveWorkedWith'].str.split(';')
df_exploded = df.explode('NEWCollabToolsHaveWorkedWith').rename(columns={'NEWCollabToolsHaveWorkedWith': 'CollabTool'})
df_exploded['CollabTool'] = df_exploded['CollabTool'].str.strip()

# Step 2: Group by Age and Collaboration Tool to count frequency
grouped = df_exploded.groupby(['Age', 'CollabTool']).size().reset_index(name='Count')

# Optional: Sort age groups
age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older'
]
grouped['Age'] = pd.Categorical(grouped['Age'], categories=age_order, ordered=True)
grouped = grouped.sort_values(['Age', 'CollabTool'])

# Step 3: Bubble Plot
plt.figure(figsize=(12, 7))
scatter = plt.scatter(
    x=grouped['CollabTool'],
    y=grouped['Age'],
    s=grouped['Count'] * 150,  # scale bubble size
    alpha=0.6,
    edgecolors='black',
    c=grouped['Count'],
    cmap='cool'
)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Frequency of Tool Usage')

# Final plot tweaks
plt.title('Bubble Plot: Collaboration Tools Used by Age Group')
plt.xlabel('Collaboration Tool')
plt.ylabel('Age Group')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here

# Sample data — replace this with your actual DataFrame
df = pd.DataFrame({
    'WebframeWantToWorkWith': [
        'React.js;Vue.js', 'Django;Flask', 'React.js', 'Angular;Vue.js',
        'Django;React.js', 'Vue.js', 'Flask;Svelte'
    ],
    'JobSatPoints_6': [4, 5, 3, 4, 5, 3, 2]
})

# Step 1: Explode WebframeWantToWorkWith into individual frameworks
df = df.dropna(subset=['WebframeWantToWorkWith', 'JobSatPoints_6']).copy()
df['WebframeWantToWorkWith'] = df['WebframeWantToWorkWith'].str.split(';')
df_exploded = df.explode('WebframeWantToWorkWith').rename(columns={'WebframeWantToWorkWith': 'WebFramework'})
df_exploded['WebFramework'] = df_exploded['WebFramework'].str.strip()

# Step 2: Group by framework to get average satisfaction and count
grouped = df_exploded.groupby('WebFramework').agg(
    AvgJobSat=('JobSatPoints_6', 'mean'),
    RespondentCount=('WebFramework', 'count')
).reset_index()

# Step 3: Create the Bubble Plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(
    x=grouped['WebFramework'],
    y=grouped['AvgJobSat'],
    s=grouped['RespondentCount'] * 150,  # Bubble size
    alpha=0.6,
    edgecolors='black',
    c=grouped['AvgJobSat'],  # Color by satisfaction
    cmap='plasma'
)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Average Job Satisfaction')

# Final plot settings
plt.title('Bubble Plot: Preferred Web Frameworks vs. Job Satisfaction')
plt.xlabel('Web Framework')
plt.ylabel('Average Job Satisfaction (Points)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
